```python
# #Write a Python program to create a joinplot to describe individual distributions on the same plot between Sepal length and Sepal width.

# Note: The bivariate analogue of a histogram is known as a hexbin plot, because it shows the counts of observations that fall within hexagonal bins. This plot works best with relatively large datasets. Its available through the matplotlib plt.hexbin function and as a style in jointplot(). It looks best with a white background.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
iris = pd.read_csv("https://raw.githubusercontent.com/roholy/Iris/main/Data/iris%20(1).csv")
fig=sns.jointplot(x='SepalLengthCm', y='SepalWidthCm', kind="hex", color="red", data=iris)
plt.show()
```

```python
# Write a Python program to create a joinplot using kde to describe individual distributions on the same plot between Sepal length and Sepal width.

# Note: The kernel density estimation (kde) procedure visualize a bivariate distribution. In seaborn, this kind of plot is shown with a contour plot and is available as a style in jointplot().


import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
iris = pd.read_csv("https://raw.githubusercontent.com/roholy/Iris/main/Data/iris%20(1).csv")
fig=sns.jointplot(x='SepalLengthCm', y='SepalWidthCm', kind="kde", color='red', data=iris)  
plt.show()
```
